# Introduction
This dataset provides a huge amount of information on crop production in India ranging from several years. Data for the crops spans over a range of 14 years, from 2000 to 2014, from 33 Indian provinces [States and Union Territories] consisting of a total 646 districts from the Northern, Eastern, Western and Southern geographic regions of India. 

A total of 6 annual seasonal periods have been taken into account when constructing this dataset, namely Whole Year, Summer, Autumn, Winter, Kharif and Rabi. These are the different planting, growing and harvesting periods for the various crops of India.

•	Crops may take a whole year to grow, planted in January and harvested by the end of December. Crops like Coconut, Garlic and Sugarcane fall under this category.

•	Summer crops like pumpkin, cucumber, pepper, tomato, and some coarse cereals are sown during March and are generally harvested by the end of June. 

•	Autumn crops like lettuce, peas, onions, and spinach need cooler temperatures when compared with summers, autumn continues majorly for two months i.e. September and October. 

•	Winter crops like cauliflower, broccoli, and sesamum require the temperature to be as low as 10 degrees Celsius and thus the chilly months - November till February – are suitable for growing them

•	Rabi crops are known as winter crops. They are grown in October or November. The crops are then harvested in spring. These crops require frequent irrigation because these are grown in dry areas. Wheat, gram and barley are some of the rabi crops grown in India.

•	Unlike Rabi crops, Kharif crops require good rainfall. The output of these crops depends upon the time and amount of rainwater. Paddy, maize, bajra, jowar are a few of the Kharif crops grown in India. Harvesting season begins from the 3rd week of September to October.

Area of each harvested crop is displayed in hectares, and Production displayed is in the unit of metric tonnes.

# Creating a New Dataset for Further Analysis:

```python
# Creating zones
north_india = ['Jammu and Kashmir', 'Punjab', 'Himachal Pradesh', 'Haryana', 'Uttarakhand', 'Uttar Pradesh', 'Chandigarh']
east_india = ['Bihar', 'Odisha', 'Jharkhand', 'West Bengal']
south_india = ['Andhra Pradesh', 'Karnataka', 'Kerala' ,'Tamil Nadu', 'Telangana']
west_india = ['Rajasthan' , 'Gujarat', 'Goa','Maharashtra']
central_india = ['Madhya Pradesh', 'Chhattisgarh']
north_east_india = ['Assam', 'Sikkim', 'Nagaland', 'Meghalaya', 'Manipur', 'Mizoram', 'Tripura', 'Arunachal Pradesh']
ut_india = ['Andaman and Nicobar Islands', 'Dadra and Nagar Haveli', 'Puducherry']
```

```python
# Creating column zones in the new dataset
def get_zonal_names(row):
    if row['State_Name'].strip() in north_india:
        val = 'North Zone'
    elif row['State_Name'].strip()  in south_india:
        val = 'South Zone'
    elif row['State_Name'].strip()  in east_india:
        val = 'East Zone'
    elif row['State_Name'].strip()  in west_india:
        val = 'West Zone'
    elif row['State_Name'].strip()  in central_india:
        val = 'Central Zone'
    elif row['State_Name'].strip()  in north_east_india:
        val = 'NE Zone'
    elif row['State_Name'].strip()  in ut_india:
        val = 'Union Terr'
    else:
        val = 'No Value'
    return val

new_df = df.copy()


new_df['Zones'] = new_df.apply(get_zonal_names, axis=1)
new_df.head()
```

```python
# Zones Column
new_df.Zones.value_counts()
```

```python
# Creating crop categories
cereal = ['Rice','Maize','Wheat','Barley','Varagu','Other Cereals & Millets','Ragi','Small millets','Bajra','Jowar',
          'Paddy','Total foodgrain','Jobster']
pulses = ['Moong(Green Gram)','Urad','Arhar/Tur','Peas & beans','Masoor','Other Kharif pulses','other misc. pulses','Ricebean (nagadal)',
          'Rajmash Kholar','Lentil','Samai','Blackgram','Korra','Cowpea(Lobia)','Other  Rabi pulses','Other Kharif pulses',
          'Peas & beans (Pulses)','Pulses total','Gram']
fruits = ['Peach','Apple','Litchi','Pear','Plums','Ber','Sapota','Lemon','Pome Granet','Other Citrus Fruit',
          'Water Melon','Jack Fruit','Grapes','Pineapple','Orange','Pome Fruit','Citrus Fruit','Other Fresh Fruits',
          'Mango','Papaya','Coconut ','Banana','Other Dry Fruit']
beans = ['Bean','Lab-Lab','Moth','Guar seed','Soyabean','Horse-gram']
vegetables = ['Turnip','Peas  (vegetable)','Beet Root','Carrot','Yam','Ribed Guard','Ash Gourd','Pump Kin','Redish','Snak Guard',
              'Bottle Gourd','Bitter Gourd','Cucumber','Drum Stick','Cauliflower','Beans & Mutter(Vegetable)','Cabbage',
              'Bhindi','Tomato','Brinjal','Khesari','Sweet potato','Potato','Onion','Tapioca','Colocosia','Other Vegetables']
spices = ['Perilla','Ginger','Cardamom','Black pepper','Dry ginger','Garlic','Coriander','Turmeric','Dry chillies','Cond-spcs other']
fibres = ['other fibres','Kapas','Jute & mesta','Jute','Mesta','Cotton(lint)','Sannhamp']
nuts = ['Arcanut (Processed)','Atcanut (Raw)','Cashewnut Processed','Cashewnut Raw','Cashewnut','Arecanut','Groundnut']
oilseeds = ['other oilseeds','Safflower','Niger seed','Castor seed','Linseed','Sunflower','Rapeseed &Mustard','Sesamum','Oilseeds total']
commercial = ['Tobacco','Coffee','Tea','Sugarcane','Rubber']
```

```python
# Creating column Crop_Cat in new dataset
def cat_crop(crop):
    for i in cereal:
        if crop==i:
            return 'Cereal'
    for i in pulses:
        if crop==i:
            return 'Pulses'
    for i in fruits:
        if crop==i:
            return 'Fruits'
    for i in beans:
        if crop==i:
            return 'Beans'
    for i in vegetables:
        if crop==i:
            return 'Vegetables'
    for i in spices:
        if crop==i:
            return 'Spices'
    for i in fibres:
        if crop==i:
            return 'Fibres'
    for i in nuts:
        if crop==i:
            return 'Nuts'
    for i in oilseeds:
        if crop==i:
            return 'Oilseeds'
    for i in commercial:
        if crop==i:
            return 'Commercial'

crop=new_df['Crop']
new_df['Crop_Cat']=new_df['Crop'].apply(cat_crop)
```

# Implementing Feature Engineering

Some observations are far too voluminous in their raw state to be modelled by
predictive modelling algorithms directly. Common examples include image,
audio, and textual data, but could just as easily include tabular data with
millions of attributes. Feature extraction is a process of automatically reducing
the dimensionality of these types of observations into a much smaller set that
can be modelled. In our case, Principal Component Analysis or PCA may be
employed.

## 1. Importing required libraries

In addition to the 4 other libraries we have imported, we import **StandardScaler** from the **sklearn.preprocessing** library. This is required to scaled our feature vectors to all having the same scale.

```python
from sklearn.preprocessing import StandardScaler
```

## 2. Extracting the required data and splitting into feature vectors

We split our dataset into 3 features: **Area, Production and Crop Year**
Here, our a values consist of the **Area** and **Production** feature vectors,
while our b values will consist of **Crop Year** feature vectors.

```python
df = pd.read_csv("UE19CS306A_5_T24.csv")
df.dropna(subset=["Production"],axis=0,inplace=True)
a = df.loc[:, ['Area','Production']] 
b = df.loc[:, ['Crop_Year']]
```

## 3. Scaling feature vectors using StandScaler()

```python
a_std = StandardScaler().fit_transform(a)
print(a_std)
```

## 4. Forming a covariance matrix of feature vectors

```python
#features are columns from a_std.T
features = a_std.T
cov_matrix = np.cov(features)
print(cov_matrix)
```

## 5. Obtaining Eigen Values and Eigen Vectors from Covariance Matrix

### Eigenvector

It is a non-zero vector that stays parallel after matrix multiplication.Basically, it is variance focused approach seeking to reproduce total variance and correlation with all components. The principal components are basically the linear combinations of the original variables weighted by their contribution to explain the variance in a particular orthogonal dimension.

### Eigenvalue

It is basically known as characteristic roots. It basically measures the variance in all variables which is accounted for by that factor. The ratio of eigenvalues is the ratio of explanatory importance of the factors with respect to the variables.In simple words, it measures the amount of variance in the total given database accounted by the factor. We can calculate the factor’s eigenvalue as the sum of its squared factor loading for all the variables.


```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
print('Eigen Values\n%s' %eig_vals)
print('\nEigen Vectors\n%s' %eig_vecs)
```

We reduce dimensions to 1

```python
eig_vals[0] / sum(eig_vals) #variance of 47% 
```

## 6. Projecting Datapoints onto selected Eigenvector

Using the important values, obtained , we project datapoints of Crop_year onto the Eigen vector to form our 1-D PCA plot

```python
projected_a = a_std.dot(eig_vecs.T[0])
result = pd.DataFrame(projected_a, columns=['PC1'])
result['y-axis'] = 0.0 #plotting in 1D, so we only plot along x-axis
result['Crop_Year'] = b
result.head(22) #first 22 values
```

```python
sns.lmplot(x='PC1', y='y-axis', data=result, fit_reg = False,
            scatter_kws={"s":10},
            hue="Crop_Year",
            height=10, aspect=15/10)

plt.title('Result')
```

# Choosing an ML model

To decide on the optimum machine learning model for our dataset, we have to
take into consideration its many various aspects.

• Our dataset consists of a vast higher number of observations as
compared to features, so algorithms with high bias/low variance like
Linear regression, Naïve Bayes, or Linear SVM are **unsuitable**, while low
bias/high variance algorithms like KNN, Decision trees, or kernel SVM
are **preferred**

• Since our dataset is not exactly linear, we require other algorithms which
can handle high dimensional and complex data structures. Therefore, 
algorithms which assume linearity like logistic regression and linear
support vector machines are **unsuitable**, while algorithms which can
handle complex data like kernel SVM, Random forest, Neural Nets are
**preferred**.

• Crop cultivation is integral to sustaining the population of a country,
therefore accuracy of the ML is very important. However, high accuracy
requires more training time. Our dataset is updated every season,
around 3-3.5 months, which is sufficiently long. Therefore, high accuracy
models with long training times like SVM or Neural Networks are
**preferred**.

From the the considerations above, the most suitable ML model is a **Support
Vector Machine using the kernel method** for highly complex data.
